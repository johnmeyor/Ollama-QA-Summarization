# Title: Deploying Ollama LLM with Colab-Xterm and LangChain in Google Colab

**Description:** This Colab notebook code demonstrates how to set up a terminal interface using the colab-xterm extension, install and integrate LangChain (including the community extensions), and deploy the Ollama LLM on Google Colab. The code performs the following steps:

**1. Terminal Setup with colab-xterm:**


*   Installs the colab-xterm package, which adds an xterm terminal to your Colab environment.

*   Loads the extension (%load_ext colabxterm) so you can run shell commands interactively within Colab.


**2. Dependency Installation:**

Silently installs langchain and langchain_community via pip, ensuring that the necessary modules for working with language models are available.

**3. Launching xterm:**

Invokes the %xterm magic command to open an interactive shell session inside Colab, which is useful for running CLI tools directly.

**4. Installing and Running Ollama:**

*  Uses a curl command to download and run the Ollama installation script from its official website.

*  Executes commands (ollama serve & ollama pull llama3) to start the Ollama service and pull a specific model ("llama3"). This setup allows the Colab instance to serve as a host for the Ollama model.

**5. Integrating Ollama with LangChain:**

*  Imports the Ollama module from langchain_community.llms.

*  Initializes an instance of the Ollama LLM using the "llama3" model.

*  Invokes the model to generate a response to a prompt ("What is the capital of Florida?") and prints the returned output.

```python
# Import Ollama module from Langchain
from langchain_community.llms import Ollama

# Initialize an instance of the Ollama model
llm = Ollama(model="llama3")
# Invoke the model to generate responses
response = llm.invoke("What is the capital of Florida?")
print(response)
```
